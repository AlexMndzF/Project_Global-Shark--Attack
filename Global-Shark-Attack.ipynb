## 1.- Importamos el data set 
El data set se encuentra en la carpeta input, por lo que en la funcion read_csv() le damos el path en el que se encuentra el fichero

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import re
from src.functions import changeCharacterSex, changeFatalValue, removefinalcharacters,ChangeType
data = pd.read_csv("input/GSAF5.csv", encoding = "ISO-8859-1")
display(data)
```

## 2.- Comprobamos los valores nulos del dataframe creado

```python
print(data.isnull().sum())
```

Eliminamos las columnas Unmamed: 22 y Unnamed:23 por tener la primera un dato y la segunda dos datos y no hacer referencia a que datos son

```python
data = data[['Case Number', 'Date', 'Year', 'Type', 'Country', 'Area', 'Location',
       'Activity', 'Name', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)', 'Time',
       'Species ', 'Investigator or Source', 'pdf', 'href formula', 'href',
       'Case Number.1', 'Case Number.2', 'original order']]
```

```python
print(data['Type'].value_counts()) 
print(data['Fatal (Y/N)'].value_counts())
data['Sex '] = data['Sex '].fillna('UNKNOWN')
print(data['Sex '].value_counts())
```

```python
data["Sex "] = data["Sex "].apply(changeCharacterSex)
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].fillna('UNKNOWN')
data['Fatal (Y/N)'] = data['Fatal (Y/N)'].apply(changeFatalValue)
data['Type'] = data['Type'].apply(ChangeType)
data['Country'] = data['Country'].fillna('UNKNOWN')
data['Country']=data['Country'].apply(removefinalcharacters) 
print(set(data["Sex "]),set(data['Fatal (Y/N)']),set(data['Type']))
```

```python
data1 = data[['Case Number', 'Date', 'Year', 'Type', 'Country', 'Area', 'Location',
       'Activity', 'Name', 'Sex ', 'Age', 'Fatal (Y/N)', 'Time',
       'Species ']]
```

```python
#Ataques totales
data_filtered = data1[(data1['Year']>=1800)&(data1['Fatal (Y/N)']!="UNKNOWN")]
print(set(data_filtered['Year']),set(data_filtered['Fatal (Y/N)']))
#Ataques Fatales
data1_filtered = data1[(data1['Year']>=1800)&(data1['Fatal (Y/N)']=="Y")]
print(set(data1_filtered['Year']),set(data1_filtered['Fatal (Y/N)']))
#Ataques supervivientes
data2_filtered = data1[(data1['Year']>=1800)&(data1['Fatal (Y/N)']=="N")]
print(set(data2_filtered['Year']),set(data2_filtered['Fatal (Y/N)']))
```

```python
atacks_year = data_filtered.groupby('Year',as_index=False).agg({'Fatal (Y/N)': 'count'})
fatals_year = data1_filtered.groupby('Year',as_index=False).agg({'Fatal (Y/N)': 'count'})
survival_year = data2_filtered.groupby('Year',as_index=False).agg({'Fatal (Y/N)': 'count'})
```

```python
atacks_year.columns = ['Year', 'Atacks']
fatals_year.columns = ['Year', 'Fatals']
survival_year.columns = ['Year', 'Survivals']
data_clean = pd.merge(atacks_year, fatals_year, on='Year')
data_clean = pd.merge(data_clean, survival_year, on='Year')
display(data_clean.head())
```

```python
mean = (sum(data_clean['Atacks'])/len(data_clean['Atacks']))
# axis x, axis y
y = data_clean['Atacks']
x = data_clean['Year']

# plot 
plt.plot(x, y)
plt.axhline(y=mean, linewidth=1, color='b')
plt.xlabel('Years')
plt.ylabel('atacks year')


mean = (sum(data_clean['Fatals'])/len(data_clean['Fatals']))
# axis x, axis y
y = data_clean['Fatals']
x = data_clean['Year']

# plot
plt.plot(x, y)
plt.axhline(y=mean, linewidth=1, color='r')
plt.xlabel('Years')
plt.ylabel('Fatals')
plt.title('Fatal per year')

mean = (sum(data_clean['Survivals'])/len(data_clean['Survivals']))
# axis x, axis y
y = data_clean['Survivals']
x = data_clean['Year']

# plot
plt.plot(x, y)
plt.axhline(y=mean, linewidth=1, color='g')
plt.xlabel('Years')
plt.ylabel('Qty')
plt.title('Atacks,Survivals,Fatals per year')


print('Blue = Atacks \nGreen = Survivals \nOrange = Fatals')
```
